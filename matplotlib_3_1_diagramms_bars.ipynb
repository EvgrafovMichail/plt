## 3.1 Работа со столбчатыми диаграммами. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1. plt.bar()

**plt.bar(x, hight, widht, color, edgecolor, linewidth, tick_label, label)** - столбчатая диограмма: *x* - деления, на каторых будут расположены столбцы - список, *hight* - список высот столбцов, каждой позиции из х в соответствие ставится высота из hight, *width* - ширина столбцов, число или список(аналогично hight, в общем), *color* - цвет столбцов, *edgecolor* - цвет контура столбцов, *linewidth* - толщина контура - дробное число от 0 до 1, *tick_label* - массив текстовых меток для каждого столбца. 

#### Пример №1:

```python
fig = plt.figure(figsize=(6.4 * 1.5, 4.8 * 1.5))

plt.bar(np.linspace(-10, 10, 21), np.linspace(1, 21, 21), width=0.8, color='c',
       edgecolor='k', linewidth=1, tick_label=['A' for i in range(21)])

plt.show()
```

#### Пример №2:

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men, women = [20, 34, 30, 35, 27], [25, 32, 34, 20, 25]

x = np.arange(len(labels))

width = 0.4

fig, ax = plt.subplots(figsize=(6.4 * 1.5, 4.8 * 1.5))

diagramm1 = ax.bar(x - width / 2, men, width, color='c', edgecolor='k', 
                  label='Men')
diagramm2 = ax.bar(x + width / 2, women, width, color='m', edgecolor='k',
                  label='Women')

ax.set_ylabel('Sorce', fontsize=13, fontweight='bold')
ax.set_title('Scorces by group and gender', fontsize=20, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=13, fontweight='bold')

ax.legend()
ax.grid()

plt.show()
```

### 2. plt.barh()

**plt.barh()** - то же самое, что и просто plt.bar(), только столбцы откладываются на вертикальной шкале OY. 

#### Пример №1:

```python
fig = plt.figure(figsize=(6.4 * 1.5, 4.8 * 1.5))

plt.barh(np.linspace(0, 10, 11), np.linspace(1, 10, 11), height=0.8, color='c',
        edgecolor='k')
```
