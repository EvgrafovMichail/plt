```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Опыт №1. Зависимость интенсивности от луча поворота анализатора.

```python
df = pd.read_csv('232_1.csv', sep=';')
df.plot(x='angle', y='intensity', figsize=(6.4 * 2, 4.8 * 1.5), grid=True,
       ylim=(0, df['intensity'].max()), xlim=(0, 180), title=r'$I_0cos^2(\alpha)$')

plt.show()
```

### Опыт №2. Зависимость интенсивности от угла поляризации при фиксированном угле падения. 

$Угол$  $падения$  $\Theta = 50$

```python
df = pd.read_csv('232_2.csv', sep=';')
df['angle'] = df['angle'] + 275
df.plot(x='angle', y='intensity', figsize=(6.4 * 2, 4.8 * 1.5), grid=True,
       xlim=(df['angle'].min(), df['angle'].max()))

plt.show()
```
