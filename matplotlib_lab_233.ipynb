### Опыт №2. Распределение напряжённости поля с диэлектриком. 

**Экспериментальные данные**

```python
distance = np.array([12, 12.7, 13, 14, 15, 16, 17,
                    18, 19, 20, 21, 21.15, 22])
amplitude = np.array([1.75, 1.55, 1.8, 2.55, 3.4, 4,
                     4.2, 3.8, 3.05, 2.2, 1.6, 1.55, 1.8])
```

**Графическое представление**

Расчёт длины волны производим по формуле: $\lambda_v = \frac{\lambda_0}{\sqrt{1 - (\frac{\lambda_0}{2a})^2}}$, где $\lambda_0 = \frac{2\pi c}{\omega}$

**Расчёт длины волны**

```python
c = 3 * 10 ** 8                             #скорость света в вакууме
omega = 27.09 * 10 ** 9                     #круговая частота волны

lambda0 =  c / omega * 1000

print(round(lambda0, 2))
```

$$\lambda_0 = 11.07 mm$$ 

**Расчёт длины волны в волноводе**

```python
a = 7.2                         
sqrt = np.sqrt(1 - (lambda0 / (2 * a)) ** 2)
lambdav = lambda0 / sqrt                     #длина волны в волноводе

print(round(lambdav, 2))
```

$$\lambda_v = 17.33mm$$

### Расчёт коэффициента отражения:

Теоретические выкладки:

$\rho = \frac{w_2 - w_1}{w_2 + w_1}$,    $ksw = \frac{1 + \rho}{1 - \rho}$,    $ksw = \sqrt{\frac{A_{max}}{A_{min}}}$

Откуда:

$ksw - ksw \times \rho = 1 + \rho$ 

$ksw - 1 = (ksw + 1) \times \rho$

$\rho = \frac{ksw - 1}{ksw + 1}$

**При замыкании(проводящая пластина)** $\rho = -1$

**Для диэлектрика:**

```python
ksw = np.sqrt(amplitude.max() / amplitude.min())
rho = (ksw - 1) / (ksw + 1)

print(round(rho, 3))
```

$\rho = 0.244$

### Рассчёт диэлектрической проницаемости для диэлектрика.

Теоретические выкладки:

$\rho = \frac{w_2 - w_1}{w_2 + w_1}$

$\rho = \frac{\sqrt{\frac{\mu_2}{\epsilon_2}} - \sqrt{\frac{\mu_1}{\epsilon_1}}}{\sqrt{\frac{\mu_2}{\epsilon_2}} + \sqrt{\frac{\mu_1}{\epsilon_1}}}$

$\mu_2 = \epsilon_2 = \mu_1 = 1$

$\rho = \frac{1 - \sqrt{\frac{1}{\epsilon_1}}}{1 + \sqrt{\frac{1}{\epsilon_1}}}$

$\rho + \rho \times \sqrt{\frac{1}{\epsilon_1}} = 1 - \sqrt{\frac{1}{\epsilon_1}}$

$(1 + \rho) \times \sqrt{\frac{1}{\epsilon_1}} = 1 - \rho$

$\sqrt{\frac{1}{\epsilon_1}} = \frac{1 - \rho}{(1 + \rho)}$

$\frac{1}{\epsilon_1} = (\frac{1 - \rho}{(1 + \rho)})^2$

$\epsilon_1 = (\frac{1 + \rho}{1 - \rho})^2$

```python
epsilon = ((1 + rho) / (1 - rho)) ** 2

print(round(epsilon, 1))
```

$$\epsilon = 2.7$$

## Пункт 2. Диаграмма направленности.

**Экспериментальные данные:**

```python
angle = np.arange(0, 70, 10)
amplitude = np.array([3.6, 2.6, 1.15, 0.25,
                     0.05, 0.025, 0])
```

**Построение в полярных координатах:**

```python
fig =plt.figure(figsize=(6.4 * 1.5, 4.8 * 1.5))
ax = fig.add_axes([0, 0, 1, 1], polar = True)

ax.scatter(angle / 180 * np.pi, amplitude, color='b')
ax.scatter((-1) * angle / 180 * np.pi, amplitude, color='b')
ax.set_rlim(0, amplitude.max())

plt.show()
```

**Построение в Декартовых прямоугольных координатах:**

```python
fig, ax = plt.subplots(figsize=(6.4 * 1.5, 4.8 * 1.5))

ax.scatter(angle, amplitude, color='b')
ax.scatter(-angle, amplitude, color='b')
ax.set_xlim(-60, 60)
ax.set_xticks(np.arange(-60, 70, 10))
ax.set_ylim(0)

ax.grid()

plt.show()
```
