# Matplotlib#1. START

```python
import matplotlib.pyplot as plt
```

## 1. Освоение примитивов. 

**figure()** - функция для создания рисунка. Рисунок - это поле, к которому будут крепиться экземпляры **axes**, т.е. поля для рисования. figure - основа всего графического представления. Рисунок задаётся в относительных координатах от нуля до одного. 

**scatter()** - маркировка указанной точки. Среди необязательный параметров есть параметр **с**, изменение которого привоит к изменению цвета маркера. 

**show()** - отображение нарисованного.

**savefig()** - сохранение фигуры в файл с некоторым рассширением(jpg, png, pdf). Если в запущенном скрипте присутствует несколько рисунков, то savefig() сохранит последний из них. 

**add_axes([left, bottom, weight, hight], polar=False)** - добавление области для рисования. Первый аргумент - относительная координата левого нижнего угла для установки области рисования на figure, а также размеры создаваемой области. **polar** - флаг, для интерпретации области рисования как полярной. По умолчанию polar=False.  

*Замечание*: Все новые графические объекты наносятся поверх предыдущих, если не выбрано новое активное поле для рисований!

#### Пример№1: 

```python
fig = plt.figure() #create object 'figure'

print(fig.axes) #list of current painting zones
print(type(fig)) #fig's type

plt.scatter(1.0, 1.0, c='red')
plt.show()
```

#### Пример№2:

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
plt.scatter(1.0, 1.0)

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1], polar=True)
plt.scatter(0., 0.5, c='red')

plt.show()
```
