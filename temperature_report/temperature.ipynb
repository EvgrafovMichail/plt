# Отчёт по болезни №1. Ночь 2021.09.29 - 2021.09.30

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Константы.

```python
lenght = 12.8
width = 9.6
```

## Вспомогательные функции.

```python
def visualize_data_from_csv(table, x_col, y_col):
    
    table.plot(x=x_col, y=y_col, figsize=(lenght, width), grid=True,
             xlim=(table[x_col].min(), table[x_col].max()), color='b')

    plt.show()
```

```python
def visualize_boxplot_from_df(df, col):
    
    df.boxplot(column=col, figsize=(lenght, width), grid=True)
    
    plt.show()
```

```python
def visualize_hist_from_df(df, col, bin_num):
    
    row = df[[col]] #.div(len(df) * (bin_num + 1))
    
    row.hist(column=col, bins=bin_num, figsize=(lenght, width), grid=True)
    
    plt.show()
```

## Считывание данных. 

Отсчёт производился в минутах, начиная от 22:00 2021.09.29. Период измерений - 10 минут.

```python
table = pd.read_csv('temperature.csv', sep=';')

visualize_data_from_csv(table, 'time', 'Temperature')
```

## Анализ вариационного ряда.

#### Получение вариационного ряда из выборки:

```python
table = table.sort_values(by=['Temperature'])
```

#### Медиана:

```python
medians = table.median()

print(medians)
```

$ med = 37.4^{\circ}C $

#### Мода:

```python
modes = table.mode()

print(modes)
```

$ mode_1 = 37.3^{\circ}C $

$ mode_2 = 37.5^{\circ}C $

#### Boxplot:

```python
visualize_boxplot_from_df(table, 'Temperature')
```

#### Гистограмма:

```python
k = int(round(np.log2(len(table))))

visualize_hist_from_df(table, 'Temperature', k)
```

*Примечание*: гистограмма не нормирована. Отражает просто количество значений в диапазоне.
