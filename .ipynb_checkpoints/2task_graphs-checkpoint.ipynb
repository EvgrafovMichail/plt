# Графики для второго задания.

```python
import matplotlib.pyplot as plt
import random as rd
import numpy as np
```

### Константы.

```python
N = 100
theta = rd.randint(15, 100)
lenght = 6.4 * 1.5
width = 4.8 * 1.5
sample = np.array([rd.uniform(0, theta) for i in range(N)])
```

### Вспомогательные функции:

```python
def calculate_shifts(arr, theta, calculate_shift):
    
    return [calculate_shift(arr[ : i + 1], theta) for i in range(len(arr))]
```

```python
def visualize_data(x, y, title, xlabel='x', ylabel='y'):
    
    fig = plt.figure(figsize=(lenght, width))
    
    ax = fig.add_axes([0, 0, 1, 1])
    
    ax.set_title(title, fontsize=15, fontweight='bold')
    ax.set_xlabel(xlabel, fontsize=11, fontweight='bold')
    ax.set_ylabel(ylabel, fontsize=11, fontweight='bold')
    ax.set_xlim(x.min(), x.max())
    
    ax.plot(x, y, color='b')
    
    ax.grid()
    
    plt.show()
```

## График №1.

Оценка: $ \theta'_1 = \frac{2}{n} \sum\limits_{i = 1}^{n}x_i$

```python
def _calculate_shift(arr, theta):
    
    return np.abs(2 * arr.sum() / len(arr) - theta)
```

```python
arr = calculate_shifts(sample, theta, _calculate_shift)

visualize_data(np.linspace(1, N, N), arr, 'Graph #1', 'n', 'shift')
```

## График №2

Оценка: $ \theta'_2 = (n + 1) x_{min}$

```python
def _calculate_shift(arr, theta):
    
    return np.abs(arr.min() * (len(arr) + 1) - theta)
```

```python
arr = calculate_shifts(sample, theta, _calculate_shift)

visualize_data(np.linspace(1, N, N), arr, 'Graph #2', 'n', 'shift')
```

## График №3

Оценка: $ \theta'_3 = \frac{n + 1}{n}x_{max} $

```python
def _calculate_shift(arr, theta):
    
    return np.abs(arr.max() * (len(arr) + 1) / len(arr) - theta)
```

```python
arr = calculate_shifts(sample, theta, _calculate_shift)

visualize_data(np.linspace(1, N, N), arr, 'Graph #3', 'n', 'shift')
```

## График №4

Оценка: $ \theta'_4 = x_{max} + x_{min} $

```python
def _calculate_shift(arr, theta):
    
    return np.abs(arr.min() + arr.max() - theta)
```

```python
arr = calculate_shifts(sample, theta, _calculate_shift)

visualize_data(np.linspace(1, N, N), arr, 'Graph #4', 'n', 'shift')
```

## График №5

Оценка: $ \theta'_5 = x_1 + \frac{1}{n - 1}\sum\limits_{i = 2}^{n} $

```python
def _calculate_shift(arr, theta):
    
    if len(arr) == 1:
        
        return arr[0]
    
    return np.abs(arr[0] + arr[1 : ].sum() / (len(arr) - 1) - theta)
```

```python
arr = calculate_shifts(sample, theta, _calculate_shift)

visualize_data(np.linspace(1, N, N), arr, 'Graph #5', 'n', 'shift')
```
