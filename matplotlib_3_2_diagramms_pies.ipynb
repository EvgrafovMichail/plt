## 3.2 Работа с круговыми диаграммами.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### plt.pie()

**plt.pie(x, explode, labels, autopct, shadow, startangel, radius)** - построение круговой диаграммы. *х* - значения, python автоматически считает их процентное соотношение, *explode* - в общем, отступ для выделения какого-то куска, задаётся дробным числом, *labels* - массив меток, каждая метка ставится в соответствие значению из массива значений, *autopct* - подпись долей, *shadow* - добавление тени к диаграмме, *startangel* - начальный угол поворота диаграммы, *radius* - значение радиуса, дробное число от нуля до 1. 

*Замечание*: В конспекте я помечаю наиболее полезные для себя, на данном этапе обучения, аргументы. Для более детального ознакомления со всеми возможными настройками круговых диаграмм можно почитать документацию: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

#### Пример №1:

```python
fig, ax = plt.subplots(figsize=(6.4 * 1.5, 4.8 * 1.5))

parts = np.array([31.2, 17.8, 8, 7, 6.1, 5.7, 24.2])
labels = ['Python', 'Java', 'JavaScript', 'C#', 'PHP', 'C/C++', 'Another']
explode = np.array([0.1] + list(np.zeros(6)))

ax.pie(parts, explode=explode, labels=labels, autopct='%1.1f%%', 
      shadow=True, startangle=90)

plt.show()
```
